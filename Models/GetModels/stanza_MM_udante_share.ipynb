```python
dir = "MM_udante"
model = "udante"

config = {
        'processors': 'tokenize,mwt,lemma,pos,depparse',
        'lang': 'la',
        'tokenize_model_path': f'./{dir}/saved_models/tokenize/la_{model}_tokenizer.pt',
        'mwt_model_path': f'./{dir}/saved_models/mwt/la_{model}_mwt_expander.pt',
        'lemma_model_path': f'./{dir}/saved_models/lemma/la_{model}_nocharlm_lemmatizer.pt',
        'pos_model_path': f'./{dir}/saved_models/pos/la_{model}_nocharlm_tagger.pt',
        'depparse_model_path': f'./{dir}/saved_models/depparse/la_{model}_nocharlm_parser.pt',
    }

nlp = stanza.Pipeline(**config)

# Note: Downloads files to /root/stanza_resources/ in Colab,
# !whoami  # ---> root
# so user's home directory actually?
```

```python
test_document = '''In Catilinam orationes Cicero consul contra Catilinam et coniurationem eius declamavit.

Prima oratio, in senatu, brevis est, sed discipuli latinae linguae saepe eam legunt quod facilior lectu est quam alia carmina et orationes. '''

doc = nlp(test_document)
```

```python
from stanza.utils.conll import CoNLL

outfile = "output.conllu"

CoNLL.write_doc2conll(doc, outfile)
```
