```python
import trankit
from trankit import Pipeline, trankit2conllu
# If you get an error about numpy not found, restart the kernel.
```

```python
!pwd
%cd trankit/MERG_Classical/
!pwd
```

```python
trankit.verify_customized_pipeline(
    category='customized-mwt', # pipeline category
    save_dir='./save_dir', # directory used for saving models in previous steps
    embedding_name='xlm-roberta-base' # embedding version that we use for training our customized pipeline, by default, it is `xlm-roberta-base`
)
```

```python
p = Pipeline(lang='customized-mwt', cache_dir='./save_dir')
```

```python
doc = '''In Catilinam orationes Cicero consul contra Catilinam et coniurationem eius declamavit.

Prima oratio, in senatu, brevis est, sed discipuli latinae linguae saepe eam legunt quod facilior lectu est quam alia carmina et orationes. '''

# perform all tasks on the input
processed_doc = p(doc)
```

```python
from pprint import pprint

pprint(processed_doc)
```

```python
conlludoc = trankit2conllu(processed_doc)
print(conlludoc)
```
